```python
import plotly.graph_objects as go
import seaborn as sns
```

Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic.age, y=titanic.fare, mode='markers'))
fig.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=tips.total_bill))
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
import plotly.express as px
fig = px.bar(tips,x= "sex",y="total_bill",color="day",pattern_shape="smoker")
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
import plotly.express as px
iris = sns.load_dataset("iris")
iris = iris[["sepal_length", "sepal_width", "petal_length", "petal_width" ]]
iris.head()
```

```python
fig = px.scatter_matrix(iris)
fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Distplot: A Distplot or distribution plot, depicts the variation in the data distribution. The distplot function in Python creates a histogram of the data, which shows the frequency of different values or ranges of values in the dataset.

```python
import plotly.express as px
tips = sns.load_dataset("tips")
fig = px.histogram(tips, x="total_bill", y="tip", color="sex", marginal="rug",hover_data=tips.columns)
fig.show()
```
